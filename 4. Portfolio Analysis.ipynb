```python
import numpy as np
import pandas as pd
import yfinance as yf
import matplotlib.pyplot as plt
```

```python
tickers = ["LMT", "CAT", "JNJ", "V"]
stocks = yf.download(tickers, start="2015-01-01")
```

```python
stocks.head()
```

```python
stocks.tail()
```

```python
stocks.info()
```

```python
normalization_stocks = (stocks["Close"] / stocks["Close"].iloc[0]*100)
print(normalization_stocks)
```

```python
normalization_stocks.plot(figsize=(12,6))
plt.title("Normalized Price Chart (Base = 100)")
plt.show()
```

```python
simple_return = stocks["Close"] / stocks["Close"].shift(1) - 1
```

```python
print(simple_return.head(10))
# Calculating simple daily returns. Simple returns are preferred over logarithmic returns when aggregating individual asset performance to calculate total portfolio returns.
```

```python
avg_daily_return = simple_return.mean()
print(avg_daily_return)
```

```python
avg_annual_return = simple_return.mean()*250
print(avg_annual_return)
```

```python
avg_annual_return
weight_portfolio_a = np.array([0.25, 0.25, 0.25, 0.25])
portfolio_a = np.dot(avg_annual_return , weight_portfolio_a)
print(f'The expected annualized return for the Equally Weighted Portfolio (Portfolio A) is: {round(portfolio_a * 100, 2)}%')
```

```python
avg_annual_return
weight_portfolio_b = np.array([0.40, 0.30, 0.20, 0.10])
portfolio_b = np.dot(avg_annual_return, weight_portfolio_b)
print(f'The expected annualized return for the Custom Weighted Portfolio (Portfolio B) is: {round(portfolio_b * 100, 2)}%')
```

```python
verify_weight_a = sum(weight_portfolio_a)
print(verify_weight_a)
verify_weight_b = sum(weight_portfolio_b)
print(verify_weight_b)
```

# Conclusion: Over the 11-year observation period, the Custom Weighted Portfolio (Portfolio B) outperformed the Equally Weighted Portfolio, yielding an annualized return of 18.49%.
